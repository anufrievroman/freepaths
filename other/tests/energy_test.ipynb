```python
import numpy as np
from scipy.constants import k, electron_volt
```

```python
energy = np.array([1,2,3,1,1,2,3,2,1,3,2,3,1,2,3,1,1,2,1,3,1,2,2,2])
time =   np.array([6,5,4,5,4,6,5,4,4,4,5,6,5,4,6,6,5,5,4,6,5,4,6,5])
```

```python
mix = np.column_stack((energy, time))
mix
```

```python
uniq, inv = np.unique(mix[:,0], return_inverse=True)
sums = np.bincount(inv, weights=mix[:,1])
counts = np.bincount(inv)
means = sums / counts
result = np.column_stack((uniq, sums, means))
```
